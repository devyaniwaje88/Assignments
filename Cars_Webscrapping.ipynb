```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
url='https://www.cars.com/shopping/results/?stock_type=cpo&makes%5B%5D=mercedes_benz&models%5B%5D=&list_price_max=&maximum_distance=20&zip='
```

```python
content=requests.get(url)
```

```python
content
```

```python
soup = BeautifulSoup(content.text, 'html.parser')
```

```python
print(soup.prettify())
```

```python
soup.find_all('div',class_='vehicle-details')
```

```python
soup.find_all('h2',class_='title')
```

```python
soup.find_all('div',class_="mileage", data_qa="mileage")
```

```python
soup.find_all('span',class_="primary-price")
```

```python
soup.find_all('div',class_='dealer-name')
```

```python
soup.find_all('span', class_="sds-rating__count")
```

```python
soup.find_all('span',class_='test1 sds-rating__link sds-button-link')
```

```python
title=[]
mileage=[]
price=[]
dealer_name=[]
rating=[]
review=[]

for i in soup.find_all('div',class_='vehicle-details'):
  try:
    title.append(i.find('h2',class_='title').text)
  except:
    title.append(np.nan)
  try:
    mileage.append(i.find('div',class_="mileage", data_qa='mileage').text)
  except:
    mileage.append(np.nan)
  try:
    price.append(i.find('span',class_="primary-price").text)
  except:
    price.append(np.nan)
  try:
    dealer_name.append(i.find('div',class_='dealer-name').text)
  except:
    dealer_name.append(np.nan)
  try:
    rating.append(i.find('span', class_="sds-rating__count").text)
  except:
    rating.append(np.nan)
  try:
    review.append(i.find('span',class_='test1 sds-rating__link sds-button-link').text)
  except:
    review.append(np.nan)

```

```python
df= pd.DataFrame({
    "model":title,
    "Mileage":mileage,
    "Price":price,
    "Dealer_Name":dealer_name,
    "Rating":rating,
    "Review":review,



})
```

```python
df.head()
```

```python
df['Dealer_Name']=df['Dealer_Name'].str.split('\n').str.get(1)
```

```python
df['Review']=df['Review'].str.split().str.get(0).str.replace('(','')
```
